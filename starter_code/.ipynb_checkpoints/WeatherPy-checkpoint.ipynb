```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy import stats
from scipy.stats import linregress
import json

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
print(weather_api_key)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
#End point url
units = "imperial"
url = 'http://api.openweathermap.org/data/2.5/weather?'
query_url = f"{url}appid={weather_api_key}&units={units}&q="
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# set up lists to hold reponse info

lat = []
lng = []
maxTemp = []
humidity = []
cloudiness = []
windSpeed = []
country = []
datetime = []
cityName = []

# Loop through the list of cities and perform a request for data on each
print ("Beginning Data Retrivial...")
index = 0


for city in cities:
    response = requests.get(query_url + city).json()
    
    try:
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        maxTemp.append(response['main']['temp_max'])
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        windSpeed.append(response['wind']['speed'])
        country.append(response['sys']['country'])
        datetime.append(response['timezone'])
        cityName.append(city)
        
        
        
        print(f"{index}: sucessfull. Processing information for : {city}")

    # Handle exceptions when information is unavailable
    except:
        # Append null values
        print((f"{index}: unsucessfull. skipping information for : {city}"))
        pass
    
    #wait for a second so you don't make too many api calls
    time.sleep(1)
    # Change index
    index = index +1
    
```

```python
weatherInfo = pd.DataFrame({ 
    'City' : cityName,
    'Cloudiness' :cloudiness,
    'Max Temp': maxTemp,
    'Latitutde': lat,
    'Longitude': lng,
    'Humidity': humidity,
    'Wind Speed': windSpeed,
    'Country': country,
    'Date': datetime 

})
weatherInfo.to_csv('cities.csv')
weatherInfo.head()
```

## Linear Regression

```python
# OPTIONAL: Create a function to create Linear Regression plots
def linregression():
    #Scatter Plot as Base graphic
    scattergraph()
    
    #Linear Regression set up vairables
    x_values = data[x]
    y_values = data[y]
    (slope, intercept, rvalue, pvalue, stderr) = (stats.linregress(x_values, y_values))
    
    # Get regression values
    regress_values = x_values * slope + intercept
    
    # Create line equation string
    line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))
    
    #Print the Regression value
    print(f"The r-squared is: {rvalue**2}")

    #Plot and position the annotation
    plt.plot(x_values,regress_values,"r-")
    plt.annotate(line_eq,((x_values.min()+x_values.max())/2,y_values.min()),fontsize=15,color="red")

```
